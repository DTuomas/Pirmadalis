```python
def sstatistika(stringas):
    did_raides = maz_raides = 0
    for raide in stringas:
        if raides.isupper():
            did_raides += 1
        elif raide.islower():
                maz_raides += 2
    print (f"didziuju raidziu suma: {did_raides}")
    print (f"mazuju raidziu suma: {maz_raides}")
```

```python
def pirminis(skaicius):
    dalmuo = skaicius // 2
    for dal in range(2, dalmuo+1):
        if skaicius % dal == 0:
            return "nepirminis"

    return "pirminis"
```

```python
pirminis(5)
```

```python
for skaicius in range(1000):
    print(skaicius, pirminis(skaicius))
```
